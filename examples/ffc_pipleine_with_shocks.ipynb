```python
import agrifoodpy
from agrifoodpy.pipeline import Pipeline
```

```python
# load the pipeline
ffc_model = Pipeline.read("ffc_pipeline_with_shocks.yaml")
```

```python
# Run the model
ffc_model.run()
```

```python
food_model = ffc_model.datablock["food"]
food_model.isel(Year=-1).fbs.plot_bars(
    show="Item_origin",
    elements=["production", "imports"],
    inverted_elements=["exports", "stock", "feed", "seed", "processing", "food", "losses", "other", "residual"]
)
```

```python
food_model["food"].fbs.plot_years(
    show="Item_origin"
)
```
